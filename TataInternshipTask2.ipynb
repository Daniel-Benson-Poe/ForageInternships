##Grammar of Graphics

EffectiveVisualizationWithGrammarOfGraphics.png

Grammar of graphics: "framework which follows a layered approach to describe and construct visualizations or graphics in a structured manner." ([Towards Data Science](https://towardsdatascience.com/a-comprehensive-guide-to-the-grammar-of-graphics-for-effective-visualization-of-multi-dimensional-1f92b4ed4149))

###The Pyramid of Grammar Graphics Components

MajorComponentsOfGrammarOfGraphics.png

Data: This is the foundation and as such should always be the starting point; idenfity your dimensions.

Aesthetics: Use the identified dimensions to confirm axes. Do you need encoding (size, shape, color, etc.) for plotting multiple dimeensions?

Scale: Is scaling needed? What scale do we need for representing multiple values/ranges?

Geometric Objects: 'geoms'. How will the data points be depicted in the visualization? Do we want a point depiction, bar depiction, line depiction, etc?

Statistics: Are statistical measures needed to help support our case? Central tendency, spread, confidence intervals, etc.

Facets: Are subplots needed to help depict data dimensions?

Coordinate system: Should the visualization be based on cartesian coordinate system or polar coordinate system?

###Examples

For the examples we'll use plotnine, "an open-source Python implementation for layered grammar of graphics based on ggplot2".

```python
from plotnine import *
from plotnine.data import mtcars

mtcars.head()
```

The mtcars dataset consists of data that was extracted from the 1974 Motor Trend US magazine, and depicts fuel consumption and 10 other attributes of automobile design and performance for 32 automobiles (1973–74 models). ([Towards Data Science](https://towardsdatascience.com/a-comprehensive-guide-to-the-grammar-of-graphics-for-effective-visualization-of-multi-dimensional-1f92b4ed4149))

2-Dimensions:

```python
# Chosen dot/scatterplot for our gemoetric object
(ggplot(mtcars,
        aes('wt', 'mpg'))
        + geom_point()
        + theme_bw())
```

We can see that car mpg has a negative correlation with car wt.

3-Dimensions:
We will continue to use dot/scatterplot and add color to help depict our third dimension.

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color='factor(gear)'))
        + geom_point()
        + theme_bw())
```

Above we can eee that lower-numbered gears tend to have higher wt and lower mpg.

4-Dimensions:

To add a fourth dimension we will use both color AND size along with our dot/scatterplot.

```python
(ggplot(mtcars,
        aes('wt', 'mpg',
            color='factor(gear)', size='cyl'))
        + geom_point()
        + theme_bw())
```

Now we can see cars with higher cylinders (cyl) tend to have lower-numbered gears and, as we saw previously, this tends to lead to higher wt and lower mpg.

Another option for depicting 4-dimensions is to use color and facets:

```python
(ggplot(mtcars,
        aes('wt', 'mpg',
            color='factor(gear)'))
        + geom_point()
        + facet_wrap('~cyl')
        + theme_bw())
```

Visualizing Data Dimensions and Statistics

```python
(ggplot(mtcars,
 aes('wt', 'mpg',
     color='factor(gear)'))
  + geom_point()
  + stat_smooth(method='lm')
  + theme_bw())
```

Here we can see linear model trends for mpg based on wt

Adding a fifth dimension:

```python
(ggplot(mtcars,
        aes('wt', 'mpg',
            color='factor(gear)',
            size='cyl'))
        + geom_point()
        + facet_wrap('~am')
        + theme_bw())
```

Above we see cars with manual transmission (facet = 1) show higher numbered gears compared to cars with automatic transmission (facet = 0). We also see that higher-numbered cylinders (cyl) tend to have automatic transmision. 

6-dimensions:

```python
(ggplot(mtcars,
        aes('wt', 'mpg',
        color='factor(gear)',
        size='cyl'))
      + geom_point()
      + facet_grid('am ~ carb')
      + theme_bw())
```

0 (automatic) and 1 (manual) are facets of the y-axis, number of carburetors (carb) are facets of the x-axis, and other dimensions similar to what we have seen before. We see that cars with higher-numbered gears tend to have manual transmission(1) and higher numbers of carburetors(carb).

##Chosing the Right Chart for Data Visualization

###Comparison Charts

Comparison-Chart.jpeg

Column Charts: 

  • Compare values across multiple categories

  • Can also show info about parts of a whole across different categories both in relative termas and absolute value.

Bar Charts:

• Only difference between columns charts and bar charts is bar charts represent values on the X-axis and categories on the Y-axis

• Used to show values across categories when duration or category text is long

• We can use stacked bar charts to compare parts of a whole and compare change over categories or time.

Line Charts:

• Used to analyze sales data, year-on-year profit, increase of person's salary over the last year, amongst many other business-related analysis.

• Used to show trends over time or categories

Scatter Plots:

• Uses numerical values along both axes

• Useful for showing correlation between data points

• Used for displaying and comparing numerical values (scientific or statistical data)

###Distribution Charts

Distributions-Chart.jpeg

Histogram:

• Used for graphing frequency over a distribution. Very useful in analytics allowing to infer many useful insights.

• Visually, all bars toucheing each other with no space between them.

Box Plot:

• Line in middle of box is the median value; 50% off the data is above median value, 50% of the data is below the median value.

• Not sways by outliers like mean.

• Within box, 25% data above median and 25% data below (50% of data within the box).

• Allows us to easily spot outliers and distribution of the plot

Box-Plot.jpeg

KDE Plot:

• Kernel Density Estimation Plot

• Smooth form of histogram

• Method for visualizing distribution of observations in dataset

• Can produce plot less cluttered and more interpretable relative to histogram.

###The Breakup of a Whole Chart

Break-up-of-a-whole-Chart.jpeg

Pie Chart:

• Represents data as a part of a whole

• Each slice represents percentage the given category occupies out of the whole.

• It's best to use a pie chart if you have less than 5 categories.

Donut Chart:

• Variant of pie chart

• Displays categories as arcs rather than slices

Stacked Bar Chart:

• Used to show relative percentage of multiple data series in stacked bar.

###Relationship Chart

Scatter Plot:

• Position of each dot on horizontal axis and vertical axis not easy to see from data alone

• Used for displaying and comparing numerical values (scientific or statistical)

Line Chart:

• Also used to find relationship between two variables.

###Trend Charts

###Line Charts

• Line Chart

• Area Chart

• Multiple-axis Line Chart

Creating Line Charts – Best Practices:

• Clearly label axes

• Remove distracting chart elements – grids, color variations, loud legends

• Zoom on y-axis if data starts above zero

• Don't compare more than 5 lines – this can clutter things too much.

###Pie Charts

• Pie Chart

• Doughnut Chart

• Semi-Circle Pie Chart

• Irregular Pie Chart

Creating Pie Charts – Best Practices

• Ensure the sem of all segments is 100

• CLEAN CLEAN CLEAN. And consistency is important. Only compare a few categories. Slices have very similar size? Use a bar or column chart instead.

• NO 3-D IMAGERY OR TILT. This makes it impossible to read.

###Bar and Column Charts

• Bar Chart

• Column Chart

• Stacked Bar Chart

• Radial Chart

• Progress Bar

Creating Bar and Column Charts – Best Practices

• Start y-axis at 0

• Label axes

• Put value labels on bars

• Avoid "rainbow effect" (too many colors). Use single color or varying shades of same color.

###Treemap

Creating a Treemap – Best Practices

• Start with clean data and clear message

• Use bright and contrasting colors making it easier to eye where one region ends and the next one begtins – you should be able to look at a treemap and easily find the largest region/s.

• Label each region with text or numbers

• Avoid clutter

###Dual Axis Chart

Column and Line Chart: Combines column chart with line chart

Dual Line Chart: Compares two line charts; can be more than two lines if needed

Dual Column Chart: Shows two sets of data displayed side by side

Multiple Axes Chart: Displays most complext version of dual axis chart.

###Area Chart

Creating an Area Chart – Best Practices

• Make it easy to read, avoiding occlusion.

• Use stacked area chart for multiple data sets and to emphasize part-to-whole relationships

• Use to look at bigger picture; Population – Line charts for showing net change in population over time, area charts for showing total population over time.

• Avoid too many datasets; use line chart instead

• Give proper context with appropriate labels and legends.

###Pyramid Chart

Creating a Pyramid Chart – Best Practices

• Clearly label subcategories – what info do you want to convey? Clearly label layers.

• Organize Subcategories – decide order and value of each section

• Organize subcategoreis by hierarchy

• CONSISTENCY! Keep section spacing even and pleasing color palette

• Keep subcategories to minimum. 

###Word Cloud

Creating a Word Cloud – Best Practices

• Provide context

• Use to show frequency – avoid to simply display complex topics like budget or healthcare crisis

• Watch word length

• Great for filtering and analyzing data

• Avoid making words too similar in size or color

### Tables

Creating a Table – Best Practices

• How will table be used? Define audience.

• Consider removing grid lines for increased readability

• Always include data source

• Numbers should be aligned to right (easier comparison). Text aligned to left or center for readability.

• Use color or formatting to draw viewer to specific values
