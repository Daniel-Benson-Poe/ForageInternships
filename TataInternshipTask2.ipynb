Grammar of graphics: "framework which follows a layered approach to describe and construct visualizations or graphics in a structured manner." ([Towards Data Science](https://towardsdatascience.com/a-comprehensive-guide-to-the-grammar-of-graphics-for-effective-visualization-of-multi-dimensional-1f92b4ed4149))

###The Pyramid of Grammar Graphics Components

MajorComponentsOfGrammarOfGraphics.png

Data: This is the foundation and as such should always be the starting point; idenfity your dimensions.

Aesthetics: Use the identified dimensions to confirm axes. Do you need encoding (size, shape, color, etc.) for plotting multiple dimeensions?

Scale: Is scaling needed? What scale do we need for representing multiple values/ranges?

Geometric Objects: 'geoms'. How will the data points be depicted in the visualization? Do we want a point depiction, bar depiction, line depiction, etc?

Statistics: Are statistical measures needed to help support our case? Central tendency, spread, confidence intervals, etc.

Facets: Are subplots needed to help depict data dimensions?

Coordinate system: Should the visualization be based on cartesian coordinate system or polar coordinate system?

###Examples

For the examples we'll use plotnine, "an open-source Python implementation for layered grammar of graphics based on ggplot2".

```python
from plotnine import *
from plotnine.data import mtcars

mtcars.head()
```

The mtcars dataset consists of data that was extracted from the 1974 Motor Trend US magazine, and depicts fuel consumption and 10 other attributes of automobile design and performance for 32 automobiles (1973–74 models). ([Towards Data Science](https://towardsdatascience.com/a-comprehensive-guide-to-the-grammar-of-graphics-for-effective-visualization-of-multi-dimensional-1f92b4ed4149))

2-Dimensions:

```python
# Chosen dot/scatterplot for our gemoetric object
(ggplot(mtcars,
        aes('wt', 'mpg'))
        + geom_point()
        + theme_bw())
```

We can see that car mpg has a negative correlation with car wt.

3-Dimensions:
We will continue to use dot/scatterplot and add color to help depict our third dimension.

```python
(ggplot(mtcars,
        aes('wt', 'mpg', color='factor(gear)'))
        + geom_point()
        + theme_bw())
```

Above we can eee that lower-numbered gears tend to have higher wt and lower mpg.

4-Dimensions:

To add a fourth dimension we will use both color AND size along with our dot/scatterplot.

```python
(ggplot(mtcars,
        aes('wt', 'mpg',
            color='factor(gear)', size='cyl'))
        + geom_point()
        + theme_bw())
```

Now we can see cars with higher cylinders (cyl) tend to have lower-numbered gears and, as we saw previously, this tends to lead to higher wt and lower mpg.

Another option for depicting 4-dimensions is to use color and facets:

```python
(ggplot(mtcars,
        aes('wt', 'mpg',
            color='factor(gear)'))
        + geom_point()
        + facet_wrap('~cyl')
        + theme_bw())
```

Visualizing Data Dimensions and Statistics

```python
(ggplot(mtcars,
 aes('wt', 'mpg',
     color='factor(gear)'))
  + geom_point()
  + stat_smooth(method='lm')
  + theme_bw())
```

Here we can see linear model trends for mpg based on wt

Adding a fifth dimension:

```python
(ggplot(mtcars,
        aes('wt', 'mpg',
            color='factor(gear)',
            size='cyl'))
        + geom_point()
        + facet_wrap('~am')
        + theme_bw())
```

Above we see cars with manual transmission (facet = 1) show higher numbered gears compared to cars with automatic transmission (facet = 0). We also see that higher-numbered cylinders (cyl) tend to have automatic transmision. 

6-dimensions:

```python
(ggplot(mtcars,
        aes('wt', 'mpg',
        color='factor(gear)',
        size='cyl'))
      + geom_point()
      + facet_grid('am ~ carb')
      + theme_bw())
```

0 (automatic) and 1 (manual) are facets of the y-axis, number of carburetors (carb) are facets of the x-axis, and other dimensions similar to what we have seen before. We see that cars with higher-numbered gears tend to have manual transmission(1) and higher numbers of carburetors(carb).
